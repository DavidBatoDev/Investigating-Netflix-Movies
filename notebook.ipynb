```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

### To gather what the data really is and what is the column for each row we can loop through the DataFrame

Start by filtering movies released in 1990
### We can use methods like loc or iloc

```python
all_shows_in_1990s = netflix_df.loc[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
movies_in_1990s = all_shows_in_1990s.loc[all_shows_in_1990s['type'] == 'Movie']
movies_in_1990s.head(5)
```
